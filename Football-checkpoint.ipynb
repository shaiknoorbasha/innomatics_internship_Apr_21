```python
# Value, wage and release clause conversion
def value_wage_conversion(Value):
    if isinstance(Value,str):
        out = Value.replace('€', '')
        if 'M' in out:
            out = float(out.replace('M', ''))*1000000
        elif 'K' in Value:
            out = float(out.replace('K', " "))*1000
        return float(out)
football['Value'] = football['Value'].apply(lambda x: value_wage_conversion(x))
football['Wage'] = football['Wage'].apply(lambda x: value_wage_conversion(x))
football['Release Clause'] = football['Release Clause'].apply(lambda x: value_wage_conversion(x))
football['Release Clause'].fillna(football['Release Clause'].mean(), inplace=True)
```

```python
# categorical columns
football['Club'].fillna('No Club', inplace=True)
football['Position'].fillna("unknown", inplace=True)
```

```python
#Joined column
def clean_date(x):
    l=[]
    if isinstance(x,str):
        l=x.split(', ')
        x=l[-1]
    return(x)

football['Joined']=football['Joined'].replace(np.nan,0)
football['Joined']=football['Joined'].apply(clean_date).astype('str')
```

```python
# Contract valid 
import datetime
football["Contract Valid Until"]=pd.to_datetime(football["Contract Valid Until"])
football["year"]=pd.DatetimeIndex(football["Contract Valid Until"]).year.astype(str)
football["Contract Valid Until"].fillna("Not available", inplace=True)
```

```python
def clean_height(x):
    l=[]
    s=1
    if isinstance(x,str):
        l=x.split("'")
        i=int(l[0])
        j=int(l[1])
        s=((i*12)+j)/12
    return(s)

football["Height"]=football["Height"].apply(clean_height)
football["Height"].fillna((football["Height"].mean()), inplace=True)
```

```python
def clean_weight(x):
    
    if isinstance(x,str):
        return(x.replace('lbs', ''))
    return(x)

football["Weight"]=football["Weight"].apply(clean_weight).astype('float')
football["Weight"].fillna((football["Weight"].mean()), inplace=True)
```

```python
# function to convert skill rating
def skillConverter(val):
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1)+int(s2)
        return val
    else:
        return val

skill_columns=['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
for col in skill_columns:
    football[col]=football[col].apply(skillConverter)
    football[col].fillna(0.0, inplace=True)
```

```python
# Number of players for each categorical varibale

def pie_chart(data, field="Nationality", percent_limit=0.5, title="Number of players"):
    title +=field
    data=data[field].value_counts().to_frame()
    total=data[field].sum()
    data["percentage"]=100*data[field]/total
    percent_limit=percent_limit
    otherdata=data[data['percentage']<percent_limit]
    others=otherdata['percentage'].sum()
    maindata=data[data['percentage']>=percent_limit]
    
    data=maindata
    other_label="Others(<" + str(percent_limit) + "% each)"
    data.loc[other_label]=pd.Series({field:otherdata[field].sum()}) # Series
    
    labels = data.index.tolist()
    datavals=data[field].tolist()
    
    trace = go.Pie(labels = labels, values=datavals) # Pie
    layout=go.Layout(title=title, height=700)
    fig=go.Figure(data=[trace], layout=layout)
    fig.show()
    
pie_chart(football, "Nationality")
pie_chart(football, "Club")
pie_chart(football, "Preferred Foot")
pie_chart(football, "Work Rate", 0.1)
pie_chart(football, "Body Type", 0.1)
pie_chart(football, "Position", 0.1)    
                         
```

```python
#Distribution of age for each players
x=football['Age']
plt.figure(figsize=(10,8))
ax=sns.countplot(x,color='#00ffff')
ax.set_xlabel(xlabel='Age of players', fontsize=10)
ax.set_title(label = "Distribution of age of the players", fontsize=20)
plt.show()
```

```python
# Distribution of Age by Potential and Overall Rating
overall = pd.DataFrame(football.groupby(['Age'])['Overall'].mean())
potential = pd.DataFrame(football.groupby(['Age'])['Potential'].mean())

merged=pd.merge(overall, potential, on='Age', how='inner')
merged['Age']=merged.index

fig, ax = plt.subplots(figsize=(6,4))
fig.show()
merged.reset_index(drop = True, inplace=True)

plt.plot('Age', 'Overall', data=merged, marker='.', color='#00ffff', lw=1, label="Overall")
plt.plot("Age", "Potential", data=merged, marker='+', color='#0000cc', lw=1, label="Potential")
plt.xlabel('Overall Rating')
plt.ylabel('Average Growth by Age')
plt.legend();
```
